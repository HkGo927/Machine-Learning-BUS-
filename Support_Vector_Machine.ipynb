```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
from sklearn. metrics import classification_report
```

```python
df = pd.read_csv('iris_new.csv')
df.head()
```

```python
feature_set = df.iloc[:, 0:4]
feature_set.head()
```

```python
target = df.iloc[:, -1]
target.head()
```

```python
train_x, test_x, train_y, test_y = train_test_split(feature_set,
                                                    target,
                                                    test_size=0.30)
```

```python
classifer = SVC()
classifer.fit(train_x, train_y)
```

```python
predicted_test_y = classifer.predict(test_x)
```

```python
acy = accuracy_score(test_y, predicted_test_y)
acy
```

```python
rp = classification_report(test_y, predicted_test_y)
print(rp)
```
