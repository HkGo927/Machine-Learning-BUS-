```python
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('o_m.csv')
df.head()
```

```python
df['result'].value_counts()
```

# **Data Exploration**

```python
sns.jointplot(data=df, x='donut', y='weight', kind='reg')
plt.show()
```

```python
sns.scatterplot(data=df, x='donut', y='weight',
                hue='result', s=80, palette='hls')
plt.show()
```

# **Model Building**

```python
train_x = df.iloc[:, 0:2].values
train_y = df.iloc[:, -1].values
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(train_x, train_y)
```

```python
micky = [[40,20]]
```

```python
prediction = model.predict(micky)
print(prediction)
```
